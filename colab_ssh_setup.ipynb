## 3. Verify your data is accessible

```
import os

DRIVE_DATA = "/content/drive/MyDrive/feline-skin-disease-detection/new_data"

# Check that the data is there
for split in ["train", "val", "test"]:
    path = os.path.join(DRIVE_DATA, split)
    if os.path.exists(path):
        classes = os.listdir(path)
        print(f"{split}: {len(classes)} classes - {classes}")
    else:
        print(f"WARNING: {path} not found! Make sure you uploaded new_data/ to Drive.")
```
